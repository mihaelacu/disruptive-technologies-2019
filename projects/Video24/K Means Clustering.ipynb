```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Introduction

To demonstrate K-Means Clustering algorihm I'll use `make_blobs` method from Scikit-Learn. I will generate random data in couple of groups and then use the algorithm to separate the data points.

```python
from sklearn.datasets import make_blobs

# 2000 points of 2 dimensional data points, in 5 groups, standard deviation of 1.2
data = make_blobs(n_samples=2000, n_features=2, centers=5, cluster_std=1.2)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

## The Model

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)
kmeans.fit(data[0])
```

The labels:

```python
kmeans.labels_[:100]
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_)
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

The colors are meaningless here. The algorithm did a pretty good job. Maybe it misclassified few points on the bottom 3 clusters, but that's expected.
